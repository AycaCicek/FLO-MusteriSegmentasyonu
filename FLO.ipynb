# Customer Segmentation wiht K-Means

Standardization

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
sc = MinMaxScaler((0, 1))
rfm = sc.fit_transform(rfm)
rfm[0:5] 
```

```python
#model kuruyoruz ama bu bir gÃ¶zetimsiz Ã¶ÄŸrenme-baÄŸÄ±mlÄ± deÄŸiÅŸken yok
kmeans = KMeans(n_clusters=4, random_state=17).fit(rfm) 
kmeans.get_params()
```

```python
kÃ¼me=kmeans.n_clusters
merkezler=kmeans.cluster_centers_
kÃ¼me_aid=kmeans.labels_
ssd=kmeans.inertia_

print(f"""
*************************************************
        KÃ¼me : {kÃ¼me}
        Merkezleri Neler: {merkezler}
        Hangi KÃ¼melere Aid: {kÃ¼me_aid}
        En yakÄ±n kÃ¼meye olan uzaklÄ±klar: {ssd}
**************************************************
   """)
```

```python
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer
```

```python
kmeans = KMeans()
elbow = KElbowVisualizer(kmeans, k=(2, 20))
elbow.fit(rfm)
elbow.show()

elbow.elbow_value_
```

```python
kmeans = KMeans(n_clusters=elbow.elbow_value_).fit(rfm)
kmeans.n_clusters
kmeans.cluster_centers_
kmeans.labels_
rfm[0:5]
```

```python
clusters_kmeans = kmeans.labels_
rfm=pd.read_sql(sql,conn)
rfm["cluster"] = clusters_kmeans #hangi mÃ¼ÅŸteri hangi kÃ¼mede
rfm.head()
```

```python
rfm["cluster"] = rfm["cluster"] + 1
rfm 
```

```python
rfm['cluster'].value_counts() 
```

```python
rfmx=rfm.groupby("cluster").agg(["count","mean","median"])
rfmx
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(rfmx)
pca_components
```

```python
pca_df = pd.DataFrame(data = pca_components, columns = ['pca1', 'pca2'])
pca_df.head()
```

```python
pca_df = pd.concat([pca_df, pd.DataFrame({'cluster': kmeans.labels_})], axis = 1)
pca_df.head()
```
